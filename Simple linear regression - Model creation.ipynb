```python
#pandas is an library and it's supports for table
import pandas as pd
```

```python
#dataset is assinged to pd.read_csv
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#call dataset variable to view output
dataset
```

```python
#independent variable is assigned to dataset inputs
independent=dataset[["YearsExperience"]]
independent
```

```python
#dependent variable is assigned to dataset output
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn is to support machine learning and then import train_test_split library to split train and test
from sklearn.model_selection import train_test_split
#x_train and x_test be the input and y_train and y-test be the output of the required dataset
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#x_train,x_test,y_train,y_test call the variable to view the train and test split 
x_train,x_test,y_train,y_test
```

```python
#By using sklearn.linear_model we import linear regression algorithm to train the dataset
from sklearn.linear_model import LinearRegression
#regressor be the variable assigned to LinearRegression() 
regressor=LinearRegression()
#regressor.fit speaks about substitution process of x_train & y_train
regressor.fit(x_train,y_train)
```

```python
#weight be the varible assigned to regressor.coef_ which helps to find slope 
weight=regressor.coef_
weight
```

```python
#bias be the variable asssigned to regressor.intercept_ which helps to find initial value
bias=regressor.intercept_
bias
```

```python
#y_pred be the variable assigned to regressor.predict(x_test) which helps to find the prediction of input test value
y_pred=regressor.predict(x_test)
```

```python
#By using sklearn.metrics we imported evalution metrics as r2_score which helps to find how well model is fitted
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#By using  pickle library  it's the process of converting a Python object into a byte stream to store it in a database 
import pickle
#to save this model we used .sav
filename="finalized_model_linear.sav
```

```python
#we used pickle.dump() function to save the database foe the deployment 
pickle.dump(regressor,open(filename,'wb'))#'wb' means write binary
```

```python
#we used pickle.load() function for the deployment process
loaded_model=pickle.load(open('finalized_model_linear.sav','rb'))#'rb' means read binary
result=loaded_model.predict([[13]])#result be the variable name assigned to loaded_model.predict([[13]]) 
```
